```python
import torch
import pandas as pd
```

```python
# bg sampled data

bg_sampled_path = "data/interim/bg_stratified_sampled.csv"
bg_sampled_df = pd.read_csv(bg_sampled_path)

bg_sampled_df.head(10)
```

```python
# for each column lets check how many values are below 1e-5
for col in bg_sampled_df.columns:
    num_below_threshold = (bg_sampled_df[col] < -1e5).sum()
    print(f"Column: {col}, Values below -1e5: {num_below_threshold}")
```

```python
# ok lets remove them from the dataframe
threshold = -1e5
bg_sampled_df_cleaned = bg_sampled_df[(bg_sampled_df >= threshold).all(axis=1)]
print("Shape after cleaning:", bg_sampled_df_cleaned.shape)
bg_sampled_df_cleaned.head(10)
```
